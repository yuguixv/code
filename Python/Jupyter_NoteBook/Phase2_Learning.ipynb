```python
# 1. 定义一个列表 (就像一个向量)
data = [10, 20, 30, 40, 50, 60, 70]

# 2. 基础取值 (索引从 0 开始！)
print("第一个数:", data[0])      # 输出 10
print("第三个数:", data[2])      # 输出 30
print("最后一个:", data[-1])     # 负数代表倒数，输出 70

# 3. 切片 (Slicing) —— 建模最核心操作！
# 格式: [开始索引 : 结束索引]  (注意：包含开始，不包含结束)
print("取第2到第4个:", data[1:4])   # 输出 [20, 30, 40] (不包含索引4的50)

# 4. 常用偷懒写法
print("前3个:", data[:3])        # 省略开头，默认从0开始
print("后3个:", data[-3:])       # 省略结尾，默认取到最后
```

```python
# 模拟一组带有“脏数据”的温度记录
# -999 代表错误数据
temperatures = [25, 28, -999, 24, 30, -999, 22]

# 创建一个空列表，用来装清洗后的数据
clean_data = []

# 开始遍历（让 t 依次代表列表里的每一个数）
for t in temperatures:
    if t != -999:           # 逻辑判断：如果 t 不等于 -999
        clean_data.append(t) # 把它装进新列表里
    else:
        print("发现脏数据，已剔除") # (可选) 打印一条日志

print("清洗结果:", clean_data)
```

```python
prices = [20, 0, 25, 30, -5, 18, 0]
clean_data = []

for p in prices:
    if p > 0:
        clean_data.append(p)

print(clean_data)
```

```python
# 定义一个函数：它接收一个数字 x，返回它的平方加 10
def magic_formula(x):
    result = x ** 2 + 10
    return result  # 把结果吐出来

# 调用函数
y1 = magic_formula(5)   # 5^2 + 10 = 35
y2 = magic_formula(10)  # 10^2 + 10 = 110

print(y1, y2)
```

```python
def calc_score(cost, performance):
    score = cost*0.4 + performance*0.6
    return score

x = calc_score(80, 90)

print(x)
```
