```python
import pandas as pd
import numpy as np  # 我们需要用 numpy 来制造一个“缺失值”

# 1. 构造字典数据 (Key是列名，Value是这一列的数据)
raw_data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Math": [85, 90, np.nan, 70, 95],     # np.nan 代表缺失值 (空)
    "English": [88, 85, 90, 60, np.nan],
    "Age": [20, 21, 20, 22, 21]
}

# 2. 变成 DataFrame
df = pd.DataFrame(raw_data)

# 3. 打印出来看看
print("--- 原始表格 ---")
print(df)
```

```python
# 1. 只看前 3 行 (快速确认数据有没有读错)
print("\n--- 前3行 ---")
print(df.head(3))

# 2. 只有上帝视角的统计摘要 (这是美赛论文里必须放的一张表！)
print("\n--- 统计摘要 ---")
print(df.describe())
```

```python
# 1. 再次查看现在的表格，记住那个 NaN
print("--- 清洗前 ---")
print(df)

# 2. 计算 Math 列的平均值 (85+90+70+95)/4 = 85.0
math_mean = df["Math"].mean()

# 3. 填补空值：把所有的 NaN 替换成平均值
# inplace=True 表示直接修改原表格，而不是生成一个新的
df["Math"] = df["Math"].fillna(math_mean)

# 4. 看看清洗后的结果
print("\n--- 清洗 Math 列后 ---")
print(df)
```

```python
# 1. 简单的列选择：只看 Name 和 Math
print(df[["Name", "Math"]])

# 2. 条件筛选 (最最常用！)
# 语法逻辑：df[ 条件判断 ]
good_students = df[df["Math"] > 80]

print("\n--- 数学大于 80 分的学生 ---")
print(good_students)
```

```python
print(df)
english_mean = df["English"].mean()
df["English"] = df["English"].fillna(english_mean)

good_english_students = df[df["English"] > 80]

print(good_english_students)
```
