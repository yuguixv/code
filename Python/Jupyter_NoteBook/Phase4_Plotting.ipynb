```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 1. 设置随机种子，保证每次生成的随机数都一样
np.random.seed(42)

# 2. 生成 100 个学生的数据
df = pd.DataFrame({
    # 每天写代码的时间 (1-10小时)
    "Coding_Hours": np.random.randint(1, 10, 100),
    # 睡眠时间 (4-10小时)
    "Sleep_Hours": np.random.randint(4, 10, 100),
    # 咖啡摄入量 (0-5杯)
    "Coffee_Cups": np.random.randint(0, 5, 100)
})

# 3. 强行制造一点规律：
# 写代码越多，得分越高；但是觉睡得太少，得分会降低
df["Score"] = df["Coding_Hours"] * 8 + df["Sleep_Hours"] * 2 + np.random.randint(-5, 5, 100)

print("数据准备完毕！前5行如下：")
print(df.head())
```

```python
# 1. 计算相关系数矩阵 (Correlation Matrix)
# 这步计算每两列数据之间的关系，范围 -1 到 1
corr = df.corr()

# 2. 画布设置 (宽8，高6)
plt.figure(figsize=(8, 6))

# 3. 画热力图
# annot=True: 在格子里显示数字
# cmap='coolwarm': 红色代表正相关(一起变大)，蓝色代表负相关
# fmt='.2f': 保留两位小数
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')

# 4. 加个标题
plt.title("Correlation Analysis: What affects Score?")
plt.show()
```

```python
# 看看“代码时间”和“分数”是不是真的成正比？
plt.figure(figsize=(8, 5))

# hue="Coffee_Cups": 用颜色深浅来表示喝了多少咖啡
sns.scatterplot(data=df, x="Sleep_Hours", y="Score", hue="Coffee_Cups", palette="viridis", s=100)

plt.title("Coding vs Score")
plt.show()
```

```python
# 重画一遍
sns.heatmap(corr, annot=True, cmap='coolwarm')

# 保存为 300 DPI (印刷级清晰度) 的图片
plt.savefig("correlation_map.png", dpi=300, bbox_inches='tight')

print("图片已保存到你的文件夹！")
```
